Agrega os resultados de cada rota, formando a planilha final referente às rotas escolhidas

```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
resultados_desag = os.listdir("resultados_desagregados")
df = pd.DataFrame()

for i in tqdm(resultados_desag):
    fhand = pd.read_csv("resultados_desagregados\\"+i) 
    df_hand = pd.DataFrame(fhand.iloc[len(fhand)-1, [3,6,8]]).T # Handler (DF) extraído do csv de cada rota
    df_hand["Rota"] = i.split(".")[0] # Rota_Viagem_id adicionado ao handler
    df = pd.concat([df, df_hand]) # Handler adicionado ao arquivo final
df = df.set_index("Rota")
df["Nível de Precipitação Total (mm)"] = df["Precip. Media Acum. (mm/h)"]*(df["Tempo de Viagem Acum. (s)"]/3600)

# Salvar os resultados
df.to_excel(r"resultados.xlsx")
```
